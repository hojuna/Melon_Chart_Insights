```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# 데이터 불러오기
df = pd.read_csv(r"C:\Users\hojun\2024_10_10_project\data\token_data\total_token_kr_song_chart_data.csv")
```

```python
#영어 가사로 인한 결측치 제거
df['lyric_tokens']=df['lyric_tokens'].fillna('')
```

```python
#wordcloud를 위해 문자열로 변환
text = ' '.join(df['lyric_tokens'].values)
```

```python
df['Genre'].value_counts().head(5)
```

```python
def wordcloud():
    # 워드클라우드 생성
    wordcloud = WordCloud(font_path='malgun.ttf', 
                        background_color='white', 
                        max_words=100,
                        width=800, 
                        height=800).generate(text)

    # 워드클라우드 시각화
    plt.figure(figsize=(8, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')  # 축 제거
    plt.show()

#모든 음악의 워드클라우드 시각화
wordcloud()
```

```python
#발라드 장르의 워드 클라우드 시각

genre='발라드'
text = ' '.join(df.loc[df['Genre']==genre]['lyric_tokens'].values)

wordcloud()
```
